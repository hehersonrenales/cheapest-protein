```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filepath = ('/content/drive/MyDrive/Colab Notebooks/Protein Costs.xlsx')
df = pd.read_excel(filepath)
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
# Dropping 'Notes' column as it seems irrelevant to our visualization
df = df.drop(columns=['Notes'])
```

```python
df.value_counts('Protein Source')
```

```python
protein_price = df.groupby('Protein Source')['Cost per 20 Grams of Protein (US Dollars)'].mean()
plt.bar(protein_price.index, protein_price.values);
plt.xlabel('Protein Source')
plt.ylabel('Cost per 20 Grams of Protein (US Dollars)')
plt.title('Cheapest Way to Get Protein')
plt.xticks(rotation=75);
```
